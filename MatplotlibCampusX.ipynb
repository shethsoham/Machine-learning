```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# **Pie Chart**

```python
areas = ['Marketing','HR', 'Sales','Customer Support','Development']
budget = [2.5,4,10,53,12]

plt.pie(budget,labels = areas, autopct = '%0.1F%%')
plt.show()
```

# **Scatter**

```python
x = [10,20,30,40,50,60,70,80]
y = [70,80,90,10,20,30,20,30]

plt.scatter(x,y)
plt.xlabel('Age')
plt.ylabel('Salary')
plt.title('Age Vs Salary')
```

# **Plot Function**

```python
product_id = [1,2,3,4,5,6,7,8]
online_price = [100,200,300,400,500,600,700,800]
offline_price = [101,202,301,402,501,602,701,802]
chor_price = [105,206,305,406,505,606,705,806]

plt.plot(product_id,online_price,color='red',marker='D', markersize=10,label='Offline Price')
plt.plot(product_id,offline_price,color='yellow',marker='+', markersize=10,label='Online Price')
plt.plot(product_id,chor_price,color='green',marker='.', markersize=10,label='Chor Bazaar Price')
plt.xlabel('Product id')
plt.ylabel('Price')
plt.title('Product id Vs Price')
plt.legend()
```
