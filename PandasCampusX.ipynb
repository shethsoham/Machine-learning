# **Merge Function**

```python
import pandas as pd

left = pd.DataFrame({'id':[1,2,3], 'Name':['Vrushali','Soham','Sushi']})
right = pd.DataFrame({'id':[1,2,3], 'Name':['Prathamesh','Sami','Piyush']})


print(left,'\n',right)
```

```python
pd.merge(left,right,on='id')
```

# **Pivot Table**

```python
mask = delivery['batsman_runs'] == 6
six = delivery[mask]
six.shape
```

```python
pt = six.pivot_table(index = 'over', columns='batting_team', values='batsman_runs', aggfunc='count')
```

```python
import seaborn as sns
sns.heatmap(pt)
```
